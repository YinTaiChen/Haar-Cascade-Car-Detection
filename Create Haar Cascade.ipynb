```python
import urllib.request
import cv2
import numpy as np
import os
```

```python
def store_neg_images():
    neg_images_link = 'http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n04096066'
    neg_image_urls = urllib.request.urlopen(neg_images_link).read().decode()
    
    if not os.path.exists('neg'):
        os.makedirs('neg')
        
    pic_num = 1
    
    for i in neg_image_urls.split('\n'):
        try:
            print(i)
            urllib.request.urlretrieve(i, "neg/"+str(pic_num)+'.jpg')
            img = cv2.imread("neg/"+str(pic_num)+'.jpg', cv2.IMREAD_GRAYSCALE)
            resized_img = cv2.resize(img, (100,100))
            cv2.imwrite("neg/"+str(pic_num)+'.jpg', resized_img)
            pic_num += 1
            
        except Exception as e:
            print(str(e))
```

```python
store_neg_images()
```

```python
def store_pos_images():
    pos_images_link = 'http://image-net.org/api/text/imagenet.synset.geturls?wnid=n02958343'
    pos_image_urls = urllib.request.urlopen(pos_images_link).read().decode()
    
    if not os.path.exists('pos'):
        os.makedirs('pos')
        
    pic_num = 1
    
    for i in pos_image_urls.split('\n'):
        try:
            print(i)
            urllib.request.urlretrieve(i, "pos/"+str(pic_num)+'.jpg')
            img = cv2.imread("pos/"+str(pic_num)+'.jpg', cv2.IMREAD_GRAYSCALE)
            resized_img = cv2.resize(img, (50, 50))
            cv2.imwrite("pos/"+str(pic_num)+'.jpg', resized_img)
            pic_num += 1
            
        except Exception as e:
            print(str(e))
```

```python
store_pos_images()
```

```python
def find_neg_uglies():
    file_type = 'neg'
    for img in os.listdir(file_type):
        for u in os.listdir('uglies'):
            try:
                current_image_path = str(file_type)+'/'+str(img)
                ugly = cv2.imread('uglies/'+str(u))
                question = cv2.imread(current_image_path)
                
                if ugly.shape == question.shape and not(np.bitwise_xor(ugly, question).any()):
                    print('Found one, which is: ', current_image_path)
                    os.remove(current_image_path)
                    
            except Exception as e:
                print(str(e))
```

```python
find_neg_uglies()
```

```python
def find_pos_uglies():
    file_type = 'pos'
    for img in os.listdir(file_type):
        for u in os.listdir('uglies'):
            try:
                current_image_path = str(file_type)+'/'+str(img)
                ugly = cv2.imread('uglies/'+str(u))
                question = cv2.imread(current_image_path)
                
                if ugly.shape == question.shape and not(np.bitwise_xor(ugly, question).any()):
                    print('Found one, which is: ', current_image_path)
                    os.remove(current_image_path)
                    
            except Exception as e:
                print(str(e))
```

```python
find_pos_uglies()
```

```python
def create_description_files():
    for file_type in ['neg', 'pos']:
        
        for img in os.listdir(file_type):

            if file_type == 'pos':
                line = file_type+'/'+img+' 1 0 0 50 50\n'
                with open('info.dat','a') as f:
                    f.write(line)
            elif file_type == 'neg':
                line = file_type+'/'+img+'\n'
                with open('bg.txt','a') as f:
                    f.write(line)
```

```python
create_description_files()
```
